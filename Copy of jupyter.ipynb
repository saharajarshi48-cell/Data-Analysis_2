```python
df = pd.read_csv('Telecom Customers Churn.csv')
df
```

```python
df['Total Charges'] = df['TotalCharges'].replace(' ', '0')
df['Total Charges'] = df['Total Charges'].astype('float')
```

```python
def conv(value):
  if value == 'Yes':
    return "Yes"
  else:
    return "no"
df['SeniorCitizen']= df['SeniorCitizen'].apply(conv)
```

```python
ax = sns.countplot(x='Churn', data=df)
ax.bar_label(ax.containers[0])
plt.title('Count of Customers By Churn')
plt.show()
```

```python
plt.figure(figsize= (3,4))
gb = df.groupby('Churn').agg({'Churn': 'count'})
gb
plt.title('Percentage of churned customers')
plt.pie(gb['Churn'], labels = gb.index, autopct = "%1.2f%%")
plt.show
```

```python
plt.figure(figsize = (5,4))
sns.countplot(x='gender', data=df, hue='Churn')
plt.title('Count of customers by gender')
plt.show
```

```python
count_data = df.groupby(['gender', 'Churn']).size().unstack(fill_value=0)
percent_data = count_data.div(count_data.sum(axis=1), axis=0) * 100
plt.figure(figsize=(6,5))
bottom_vals = None
for churn_status in percent_data.columns:
    plt.bar(
        percent_data.index,
        percent_data[churn_status],
        bottom=bottom_vals,
        label=churn_status
    )
    for i, val in enumerate(percent_data[churn_status]):
        y = val/2 if bottom_vals is None else bottom_vals[i] + val/2
        plt.text(i, y, f'{val:.1f}%', ha='center', va='center', fontsize=10)

    bottom_vals = percent_data[churn_status] if bottom_vals is None else bottom_vals + percent_data[churn_status]

plt.title("Gender-wise Churn Distribution (%)")
plt.ylabel("Percentage")
plt.legend(title="Churn")
plt.show()
```

```python
sns.histplot(x='tenure', data=df,bins = 72, hue = 'Churn')
plt.title('Distribution of Tenure')
plt.show()
```

```python
plt.figure(figsize = (4,4))
ax = sns.countplot(x='Contract', data=df, hue = 'Churn')
ax.bar_label(ax.containers[0])
plt.title('Count of customers by Contract')
plt.show()
```

```python
plt.figure(figsize=(15, 12))
cols = ['Partner', 'Dependents', 'PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport']
for i, col in enumerate(cols, 1):
    plt.subplot(3, 3, i)
    sns.countplot(x=df[col], hue=df['Churn'], palette='viridis')
    plt.title(f'{col} vs Churn')
    plt.xlabel('')
    plt.ylabel('Count')
    plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```
